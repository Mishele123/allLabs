```python
gender = doc["ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ (лет)"].values
salary = doc["Зарплата (долл. в час)"].values
gender
```

## Часть 1
1.1 Найдите оценки параметров линейной регрессионной модели   зависимости ЗАРПЛАТЫ от Z.

```python
x = gender.reshape((-1, 1))
model = LinearRegression()
model.fit(x, salary)
b_0_hat = model.intercept_
print('Смещение (b0):', b_0_hat) # b₀^
b_1_hat = model.coef_[0]
print('Наклон:', b_1_hat) # b₁^
print(f"y = {b_0_hat} + {b_1_hat} * x")
```

1.2. Проверьте на уровне значимости alpha, обладает ли построенная модель высокой объясняющей способностью, или же модель низкого качества.

```python
rss = 0
ess = 0 
avg = np.average(salary)

for i in range(len(gender)):
    rss += ((b_0_hat + b_1_hat * x[i][0]) - avg) ** 2
    
for i in range(len(gender)):
    ess += (salary[i] - (b_0_hat + b_1_hat * x[i][0]))** 2
            
f1 = rss * (len(gender)-3-1)/(3*ess)
p_value = (1-stats.f.cdf(f1, dfn=3, dfd=len(gender)-3-1))
print("F-тест")
print("Статистика:",f1)
print("p-value:",p_value)
if p_value < alpha:
    print('Модель обладает высокой объясняющей способностью')
else:
    print('Модель обладает низкой объясняющей способностью')
```

1.3. Найдите оценки параметров линейной модели зависимости ЗАРПЛАТЫ от СТАЖА РАБОТЫ и ВРЕМЕНИ,ЗАТРАЧЕННОГО НА ОБРАЗОВАНИЕ.

```python
X = sm.add_constant(doc[['СТАЖ РАБОТЫ (лет)', 'ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ (лет)']])
X.columns = ['const', 'СТАЖ РАБОТЫ', 'ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ (лет)']
y = doc['Зарплата (долл. в час)']
# Построение модели
model = sm.OLS(y, X).fit()
# Вывод результатов
print(model.summary())
print("\nРегрессионная модель, прогнозирующая «Зарплата (долл. в час)»\n на основе «СТАЖ РАБОТЫ (лет)» и «ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗАНИЕ (лет)» демонстрирует умеренное соответствие (R-квадрат = 0,201).\nCтатистически значимы (p < 0,08), при этом «ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ»\n имеет наибольший прогнозируемый положительный эффект на «Зарплата (долл. в час)».\n F-статистика (65.51, p = 4.14e-26) указывает на общую значимость модели. Отклонение остатков от нормальности требует осторожности при интерпретации допущений модели.")
```

1.4. На сколько в среднем изменится ЗАРПЛАТА, если СТАЖ РАБОТЫ увеличить на 1 год, а ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ, останется неизменным?

```python
# Коэффициент при стаже работы (b1)
b_1_hat = model.params['СТАЖ РАБОТЫ']

# Вывод результата
print(f"В среднем зарплата увеличится на {b_1_hat:.2f} долларов в час, если стаж работы увеличится на 1 год, а время, затраченное на образование, останется неизменным.")
```

1.5. Какую в среднем прибавку к ЗАРПЛАТЕ дает дополнительный год обучения?

```python
# Коэффициент при времени на образование (b2)
b_2_hat = model.params['ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ (лет)']

# Вывод результата
print(f"В среднем зарплата увеличивается на {b_2_hat:.2f} долларов в час за каждый дополнительный год обучения при фиксированном стаже работы.")
```

1.6. Проверьте на уровне значимости alpha, обладает ли построенная двухфакторная модель высокой объясняющей способностью, или же модель низкого качества.

```python
# Извлечение значений F-статистики и p-value из модели
f_stat = model.fvalue
p_value = model.f_pvalue

# Вывод результатов
print(f"F-статистика: {f_stat:.2f}")
print(f"p-value: {p_value:.2e}")

# Проверка гипотезы
if p_value < alpha:
    print("Модель обладает высокой объясняющей способностью на уровне значимости α.")
else:
    print("Модель обладает низкой объясняющей способностью на уровне значимости α.")
```

1.7. С надежностью gamma постройте доверительные интервалы для теоретических коэффициентов регрессии модели.

```python
# Уровень доверия
confidence_level = gamma

# Построение доверительных интервалов
conf_intervals = model.conf_int(alpha=1-confidence_level)

# Добавление имен коэффициентов
conf_intervals.columns = ['Нижняя граница', 'Верхняя граница']
conf_intervals.index = ['const', 'СТАЖ РАБОТЫ', 'ВРЕМЯ, ЗАТРАЧЕННОЕ НА ОБРАЗОВАНИЕ']

print("Доверительные интервалы для коэффициентов регрессии:")
print(conf_intervals)
```
