### 1. Системы множеств. Вероятностное пространство

**Определение 1.1**
Элементы $\omega \in \Omega$ называют *элементарными исходами* некоторого эксперимента, а $\Omega$ —  *пространством элементарных исходов*

Подмножество $A \subset Ω$ назовем **событием**, произошедшим, если эксперимент закончился одним из элементарных исходов $\omega$, входящих в множество A. 

**Определение 1.2**
Система подмножеств A, для которых выполнен $\text{ряд условий}$ называется *сигма-алгеброй* и обозначается $\mathscr{F}$

  Первые два определения задают два параметра тройки, из которой складывается вероятностное пространство. 

**Определение 1.3**
Функция $P:\mathscr{F}\to\mathbb{R}$ называется *вероятностностной мерой* или *вероятностью*, если она принимает значения от нуля до единицы и $\text{аддитивна}$ по $\mathscr{F}$. 

Если хотим посчитать например вероятность события $A$, обозначим $P(A)$ и получим число от 0 до 1.

Итак, теперь если рассмотреть пространство, где задана тройка, определенная выше, то получится вероятностное пространство: 

$(\Omega, A, \mathscr{F})$


### 2. Элементарные вероятностные модели

***Пример 2.1***

Пусть мы бросаем симметричную монету, тогда у нас возможны два элементарных исхода:

>$\Omega = ${$\text{орел}, \text{решка}$}
Теоретически их может быть больше — монета зависла в воздухе, экспериментатор отказался бросать :) Этот пример — **классическая вероятностная модель**, когда исходов конечно и они равновероятны. В этом случае достаточно разделить 1 на $\text{мощность множества}$ элементарных исходов $|\Omega|$, чтобы посчитать вероятность наступления события "выпал орел":\$P(\text{"выпал орел"})=\frac{1}{|\Omega|}=\frac{1}{2}$

***Пример 2.2***

>При игре в дартс на Физтехе чтобы выбить максимальное число очков нужно попасть в красный круг с радиусом $r_1=2$ см. Считая, что все дротики прилетают на круглую игровую поверхность радиусом $r_2=10$ см, найти вероятность того, что дротик **не** попадет в красный круг.\Разберемся далее с **геометрической вероятностной моделью**. В этом случае $\Omega$ является $\text{область евклидова пространства}$, для которой можно задать объем (или площадь в двумерном случае). Обозначим за $A$ событие "дротик попал в красный круг"$P(A)=\frac{|A|}{|\Omega|}=\frac{S_A}{S_{\Omega}}=\frac{\pi2^2}{\pi10^2}=0.04$. Tогда вероятность того, что дротик не попал в красный круг равна $1-P(A)=1-0.04=0.96$ 

Кстати, давайте немного попрактикуемся в языке Python. В частности, реализуем функцию, которая принимает на вход радиусы $r_1, r_2$ и отдает вероятность того, что дротик **не** попадет в красный круг и добавим визуализацию.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def darts(r_1=2, r_2=10):
    """
        draws darts plot and calculates probability
        from ex 2.2
    """
    if r_2 < r_1:
        return ValueError
    
    large_circle = plt.Circle((0, 0), r_2, color='b',alpha=0.8)
    small_circle = plt.Circle((0, 0), r_1, color='r', alpha=0.8)
    
    fig, ax = plt.subplots(figsize=(5,5)) 
    ax.set_xlim(-r_2-1,r_2+1)
    ax.set_ylim(-r_2-1,r_2+1)
    ax.add_artist(large_circle)
    ax.add_artist(small_circle)
    ax.set_title('Darts')
    result_prob = 1 - (r_1/r_2)**2
    print(f'Вероятность того, что дротик не попал в красный круг равна {result_prob}')
    plt.show()
    
    
darts(r_1=10,r_2=20)
```



### 3.  Условная вероятность. Теорема Байеса

    Условная вероятность возникает в том случае, если мы хотим вычислить вероятность наступления некоторого события $A$ при условии, что событие $B$ уже наступило.


***Определение 3.1***
Условной вероятностью $A$ при условии $B$, причем $P(B)>0$, называется $P(A|B)=\frac{P(AB)}{P(B)}$. Здесь $P(AB)$ — вероятность одновременного наступления событий $A$ и $B$.

***Пример 3.1***:

>Достают две карты из колоды в 36 карт без повторения. Каков шанс, что обе карты окажутся масти треф? Рассмотрим событие **$A$="первое появление карты масти треф"** и событие **$B$="второе появление той же масти треф".** Заметим, что второе событие зависит от первого. $P(A) = \frac{9}{36}=\frac{1}{4}$ — вероятность наступления события $А$. $P(B|A) = \frac{9-1}{36-1}=\frac{8}{35}$ — вероятность наступления события $B$. $P(AB) = P(B)P(B|A) = \frac{1}{4}\frac{8}{35}=\frac{2}{35}$ — вероятность, что будут вытащены обе карты масти треф.

***Пример 3.2***

>Брошено шесть шестигранных игральных костей, с какой вероятностью на всех выпала пятерка при условии, что на первых трех выпала пятерка?$A$="на первых трех пятерка", $P(A)=\frac{1}{6^3}$,$B$="на всех костях пятерка", $P(B)=\frac{1}{6^6}$,$P(B|A)=\frac{P(AB)}{P(A)}=\frac{6^3}{6^6}=\frac{1}{216}=0.00463$


***Теорема 3.1 (Формула полной вероятности)***

Назовем *разбиением* систему событий {$A_1,..,A_n$} разбиением, если никакие два события из разбиения не пересекаются и покрывают все множество элементарных исходов $\Omega$, причем $P(A_i) > 0, \forall i \in 0..n$. Тогда для любого события $B$ верна формула:

$P(B)=\sum_{i=1}^{n}P(B|A_i)P(A_i)$

***Теорема 3.2 (формула Байеса)***

$P(A|B)=\frac{P(B|A)\,P(A)}{P(B)}$, $P(B) > 0$

Заметим, что если использовать формулу полной вероятности, то можно записать формулу Байеса в виде:

$P(A_i|B)=\frac{P(B|A_i)\,P(A_i)}{\sum_{i=1}^{n}P(B|A_i)P(A_i)}$


***Пример 3.3***

>Жители некоторых трех районов приходят в банк и хотят взять кредит. Жители первого района приходят с частотой 0.6, второго — 0.3, а третьего — с частотой 0.1. Вероятность того, что жителю первого района одобрят кредит равна 0.4, что жителю второго — 0.9, что третьего — 0.2. Известно, что последнему жителю одобрили кредит. С какой условной вероятностью это был житель третьего дома? 


>Введем обозначения: $B$='последнему жителю одобрили кредит', $A_1$='последний клиент был из первого района'; $A_2$='последний клиент был из второго района'; $A_3$='последний клиент был из третьего района'. Теперь разберемся, как следует записать события через вероятность:$P(A_1)=0.6$, $P(A_2)=0.3$, $P(A_3)=0.1$, $P(B|A_1)=0.4$, $P(B|A_2)=0.9$, $P(B|A_3)=0.2$.

>Хотим найти $P(A_3|B)=\frac{P(B|A_3) P(A_3)}{P(A)}=\frac{P(B|A_3)P(A_3)}{P(B|A_1)P(A_1)+P(B|A_2)P(A_2)+P(B|A_3)P(A_3)}=\frac{0.2\, 0.1}{0.4\,0.6+0.9\,0.3+0.2\,0.1}=0.038$

P.S. зачем изучать условную вероятность? Очень просто! В машинном обучении, например, часто приходится выбрать класс, к которому отнести некоторый объект при известных условиях. Существуют $\text{классификаторы}$ и $\text{методы}$, основанные на теореме Байеса и названные в честь этой формулы. Классическая задача — кредитный скоринг. При наборе признаков (условий) определить, стоит ли выдавать кредит человеку.

### 4. Независимость. Распределения вероятностей

**Определение 4.1**
События $A$ и $B$ назовем **независимыми**, если выполнено утверждение: 
$P(AB)=P(A)P(B)$

  На практике при подсчете вероятностей чаще всего рассматриваются события, определяемые с помощью действительных чисел.

**Определение 4.2**
Вероятностная мера $P$, заданная на $(\mathbb{R}$,$\mathscr{B}(\mathbb{R}))$, называется распределением вероятностей. Здесь $\mathscr{B}(\mathbb{R})$ — $\text{борелевская сигма-алгебра}$ — система из отрезков, (полу)интервалов на множестве действительных чисел $\mathbb{R}$.


**Определение 4.3**
Функция $F:\mathbb{R}\to[0,1]\, F(x)=P((-\infty,x])$ называется функцией распределения, соответствующей распределению вероятностей $P$.

  Проще говоря, F описывает вероятность того, что в результате эксперимента будет получено действительное число, меньшее либо равное, чем x. 
    
Заметим, что если вероятность сосредоточена не более чем в счетном числе точек, то такое распределение — **дисректное**. В случае, если вероятность распределена по более чем счетному множеству точек, то такое распределение считается **абсолютно непрерывным**. Заметим также, что не все существующие распределения делятся на эти два класса. Но мы будем рассматривать только дискретный и абсолютно 
непрерывный случаи. Введем фукнцию, характеризующую "сосредоточенность" точек, содержащих вероятность.

**Определение 4.4**
Функция $p(t), t \in \mathbb{R}$, такая, что $\int_\mathbb{R}p(t)dt=1$ и $F(x)=\int_{-\infty}^{x}p(t)dt$, называется *плотностью абсолютно непрерывного распределения* $P$, соответсвующего функции распределения $F$.

**Пояснение**: первое условие — сумма (в дисректном случае) либо интеграл (в абсолютно непрерывном случае) по всем $t$ дает единицу, то есть функция плотности содержит всю информацию о том, как, где и сколько вероятности содержится по всем возможным исходам. Второе условие эквивалентно тому, что если функция распределения показывает вероятность не превзойти значение $x$ в результате эксперимента, то плотность  — что получится ровно значение $x$. 

Для работы с распределениями отлично подходят пакеты `scipy.stats` и `numpy`.

Существует набор заранее известных распределений, для которых уже вычислены функции и известно, при каких обстоятельствах их стоит применять. Рассмотрим некоторые из них, а именно:


Начнем с важнейших примеров **дискретных** распределений. 


Важнейшие дискретные распределения:

Назовем носителем распределения множество, на котором оно определяется. Заметим, что у распределений есть параметры, которые характеризуют конкретный случай. Будем считать, что $k \in \mathbb{N}$


|Распределение|Обозначение|Носитель|Параметр|Плотность| Смысл |
|---------|---------|---------|---------|---------|---------|
|Равномерное дискретное|$R${$1,...,N$}|{$1,..,N$} |$N \in \mathbb{N}$| $p(k)=\frac{1}{N}$|Бросок $N$-гранного кубика |
|Распределение Бернулли| $Bern(p)$ | $p\in${$0,1$} |{$0,1$}|$p(0)=1-p; p(1) = p$ |Один бросок монеты |

Как видно из таблицы, каждое распределение описывает результаты эксперимента по определенному закону. И если мы видим, что эксперимент похож по сути на смысл одного из распределений, то подобрав нужные параметры, можем описать так эксперимент.


#### Равномерное дискретное распределение

Функция `randint` из `numpy` равновероятно возвращает одно из значений из множества {$1,...,N$} для каждой из n_points точек. Вы можете поменять параметры, передаваемые в функцию и посмотреть, изменится ли график.

```python
def plot_uniform(N=10000, n_points=1000):
    """
        отрисовывает распределение случайных точек по равномерному закону.
        N: параметр распределения
        n_points: число точек распределения
    """
    dst = np.random.uniform(0,N,n_points)
    plt.hist(dst,density=True,alpha=0.7, histtype='stepfilled', color='g')
    plt.title('Uniform discrete distribution')
```

```python
plot_uniform(1, 10000)
```

Что можно заметить при уменьшении числа точек? А при увеличении $N$? Попробуйте различные параметры чтобы убедиться, что максимально равномерное распределение возникает при малых значениях $N$ и больших $npoints$. Действительно, чем меньше число доступных принимаемых значений и чем больше экспериментов проведено, тем меньше **разброс** или **дисперсия**, как мы будем называть ее далее. Функция же, которая отвечает за "проведение" эксперимента называется **случайной величиной**. Но формально поговорим об этом в следующем параграфе, а теперь вернемся к рассмотрению дискретных распределений. На очереди — Бернулли.

```python
# Еще вариант. Задаем массив возможных значений
values = [1, 2, 3, 4, 5]

# Генерация 10 случайных выборок с равномерным дискретным распределением
random_samples = np.random.choice(values, size=10)
print(random_samples)
```

В этом примере из массива values выбираются 10 случайных элементов с равными шансами для каждого значения.

#### Распределение Бернулли

заметим, что здесь используем функцию `numpy.random.binomial`, которая соответсвует биномиальному распределению, по сути представляющему бросок монеты $n$ раз, таким образом, положив $n=1$, получим Бернулли

```python
# Параметры биномиального распределения
n = 10  # число испытаний
p = 0.5  # вероятность успеха

# Генерация 5 случайных чисел из биномиального распределения
random_numbers = np.random.binomial(n, p, size=5)
print(random_numbers)
```

```python
def plot_bern(p=0.5, n_points=1000):
    """
        отрисовывает распределение случайных точек по закону Бернулли.
        p: параметр распределения, равный вероятности выпадения орла 
    """
    dst = np.random.binomial(1,p,n_points)
    plt.hist(dst,density=True,histtype='stepfilled')
    plt.title('Bernoulli')
```

```python
plot_bern(0.9)
```

У нас всего два возможных значения — ноль и единица (орел и решка). И в зависимости от "симметричности" монеты один столбец будет длиннее другого

#### Распределение Пуассона

```python
# Параметры пуассоновского распределения
lam = 5  # среднее количество событий

# Генерация 10 случайных чисел из пуассоновского распределения
random_numbers = np.random.poisson(lam, size=10)
print(random_numbers)
```

```python
def plot_poisson(lmb=5, n_points=1000):
    """
        Отрисовывает распределение случайных точек по закону Пуассона.
        lambda: параметр распределения,характеризующий среднее количество
        событий за фиксированный промежуток времени
    """
    dst = np.random.poisson(lmb,n_points)
    plt.hist(dst,density=True,histtype='stepfilled', color='b')
    plt.title('Poisson')
```

#### Равномерное непрерывное распределение

```python
def plot_uniform(a=0,b=1,n_points=1000):
    """
        отрисовывает распределение случайных точек по равномерному непрерывному закону.
        a,b: начало и конец отрезка
    """
    r = sts.uniform.rvs(loc=a, scale=b-a, size=n_points)
    plt.hist(r, density=True, histtype='stepfilled', color='y')
    plt.title('Uniform')
```

#### Нормальное распределение

Его также называют **гауссовым**. С этим распределением связана важнейшая Центральная предельная теорема (ЦПТ), которая гласит, что если величина является суммой многих случайных слабо взаимозависимых величин, каждая из которых вносит малый вклад относительно общей суммы, то центрированное и нормированное распределение такой величины при достаточно большом числе слагаемых стремится к нормальному распределению. 

```python
def plot_norm(a=0,sigma=1,n_points=10000):
    """
        отрисовывает распределение случайных точек по равномерному нормальному закону.
        a — параметр, характеризующий среднее значение
        sigma — параметр, характеризующий разброс параметров (СКО)
    """
    r = sts.norm.rvs(loc=a,scale=sigma, size=n_points)
    plt.hist(r, density=True, histtype='stepfilled',alpha=0.8)
    plt.title('Normal')
```

```python
plot_norm(0,1)
```

#### Экспоненциальное распределение

Интересно, что это распределение обладает свойством отсутствия памяти. Поясним на примере. В качестве эксперимента будем засекать время ожидания пассажиром автобуса. Пусть автобусы приходят на остановку случайно, но с некоторой фиксированной средней интенсивностью. Тогда количество времени, уже затраченное пассажиром на ожидание автобуса, не влияет на время, которое ему ещё придётся прождать.

```python
def plot_exp(lmbd=0,n_points=10000):
    """
        отрисовывает распределение случайных точек по экспоненциальному закону.
        lambda — параметр, характеризующей среднее возрастание 
        интенсивности процесса за единицу времени
    """
    r = sts.expon.rvs(lmbd, size=n_points)
    plt.hist(r, density=True, histtype='stepfilled')
    plt.title('Exponential')
```

```python
plot_exp()
```

#### Числовые характеристики эмпирических распределений. Коробчатая диаграмма.

 **Определение 5.4** Медиана — это значение, которое делит ранжированный набор данных на две равные части, так что половина наблюдений меньше медианы, а другая половина — больше. Если количество наблюдений четное, медиана вычисляется как среднее двух средних значений.

**Определение 5.5** Стандартное отклонение — это мера разброса значений вокруг среднего (математического ожидания). Оно показывает, насколько значения выборки отклоняются от среднего. Вычисляется как квадратный корень из дисперсии.

 **Определение 5.6** Коэффициент ассиметрии (или асимметрия) — это мера симметричности распределения. Он показывает, в какую сторону распределение отклоняется от симметрии:
  - Положительное значение указывает на правостороннюю асимметрию (длинный хвост справа).
  - Отрицательное значение указывает на левостороннюю асимметрию (длинный хвост слева).
  - Значение 0 означает, что распределение является симметричным.

**Определение 5.7** Эксцесс — это мера "выпуклости" распределения. Он показывает, насколько "пиковым" или "плоским" является распределение по сравнению с нормальным распределением:
  - Положительное значение указывает на более острый пик (больше вероятности экстремальных значений).
  - Отрицательное значение указывает на более плоское распределение (меньше вероятности экстремальных значений).
  - Значение 0 соответствует нормальному распределению.


**Определение 5.8** Коробчатая диаграмма (или boxplot) — это графическое представление статистических данных, которое отображает распределение данных через их квартильные значения. Основные элементы коробчатой диаграммы:

 1. Коробка (Box)
- Описание: Коробка представляет собой диапазон между первым (Q1) и третьим (Q3) квартилями. Она показывает, где находится средняя половина данных.
- Смысл: Коробка делит данные на две части: нижняя часть (от Q1 до медианы) и верхняя часть (от медианы до Q3).

 2. Медиана
- Описание: Вертикальная линия внутри коробки, которая обозначает медиану (Q2) — значение, которое делит набор данных на две равные половины.
- Смысл: Медиана показывает центр распределения данных.

 3. Усы (Whiskers)
- Описание: Линии, которые выходят из коробки к минимальному и максимальному значениям, не являющимся выбросами. Обычно они простираются до 1.5 * межквартильного размаха (IQR) от Q1 и Q3.
- Смысл: Усы показывают разброс данных и позволяют увидеть, насколько далеко находятся значения от основного диапазона.

 4. Выбросы (Outliers)
- Описание: Точки, которые расположены за пределами усов (обычно это значения, превышающие 1.5 * IQR от верхнего или нижнего квартиля).
- Смысл: Выбросы указывают на аномальные значения в наборе данных, которые могут требовать дополнительного анализа.

 5. Межквартильный размах (IQR)
- Описание: Разница между третьим (Q3) и первым (Q1) квартилями (IQR = Q3 - Q1).
- Смысл: IQR показывает, насколько разбросаны центральные 50% данных.

Коробчатая диаграмма позволяет быстро оценить распределение, центральную тенденцию и разброс данных, а также выявить выбросы, что делает её полезным инструментом для анализа данных.

**Пример.** Сначала генерируем выборку из 40 нормально распределенных значений, с параметрами: математическое ожидание - 2, среднее квадратическое отклонение - 0,3. Затем строим обычную гистограмму, а потом строим коробчатую гистограмму. Далее, находим основные числовые характеристики полученного статистического распределения: математическое ожидание, медиану, дисперсию, стандартное отклонение, коэффициент ассиметрии, эксцесс.

```python

import seaborn as sns
from scipy import stats

# Генерация выборки из 40 нормально распределенных значений
mu = 2               # математическое ожидание
sigma = 0.3          # стандартное отклонение
sample_size = 40     # размер выборки

data = np.random.normal(mu, sigma, sample_size)

# Построение гистограммы
plt.figure(figsize=(12, 6))

# Гистограмма
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, density=True, alpha=0.6, color='g', label='Гистограмма')
plt.title('Гистограмма')
plt.xlabel('Значение')
plt.ylabel('Плотность')

# Коробчатая диаграмма (boxplot)
plt.subplot(1, 2, 2)
sns.boxplot(data=data)
plt.title('Коробчатая диаграмма')

plt.tight_layout()
plt.show()

# Вычисление основных числовых характеристик
mean = np.mean(data)               # Математическое ожидание
median = np.median(data)           # Медиана
variance = np.var(data)            # Дисперсия
std_dev = np.std(data)             # Стандартное отклонение
skewness = stats.skew(data)        # Коэффициент ассиметрии
kurtosis = stats.kurtosis(data)    # Эксцесс

# Вывод результатов
print(f'Математическое ожидание: {mean:.3f}')
print(f'Медиана: {median:.3f}')
print(f'Дисперсия: {variance:.3f}')
print(f'Стандартное отклонение: {std_dev:.3f}')
print(f'Коэффициент ассиметрии: {skewness:.3f}')
print(f'Эксцесс: {kurtosis:.3f}')
```

**Примечание.** Функция plt.subplot из библиотеки Matplotlib используется для создания сетки подграфиков (или осей) в одном окне графика. Это позволяет размещать несколько графиков в одном представлении. Вот основные параметры этой функции:

 1. nrows:
   - Количество строк в сетке подграфиков. Это целое число, которое определяет, на сколько строк будет разбито окно графика.

2. ncols:
   - Количество столбцов в сетке подграфиков. Это целое число, которое определяет, на сколько столбцов будет разбито окно графика.

3. index:
   - Индекс текущего подграфика, на котором будет рисоваться график. Индексация начинается с 1 и идет слева направо, сверху вниз. Например, в сетке 2x2 индексы будут: 
     - 1: верхний левый
     - 2: верхний правый
     - 3: нижний левый
     - 4: нижний правый

### Замечание. Различие в гистограмме частот и гистограмме относительных частот

В Python различие между гистограммой частот и гистограммой относительных частот можно наблюдать при построении графиков с помощью библиотеки Matplotlib. Давайте рассмотрим, как это реализуется и что значит параметр density=True.

 **Определение 5.9.** Гистограмма частот показывает абсолютное количество наблюдений, попадающих в каждый интервал (бин).

 **Определение 5.10.** Гистограмма относительных частот показывает долю наблюдений в каждом интервале относительно общего количества наблюдений.
 Параметры: Чтобы построить гистограмму относительных частот, нужно установить параметр density=True. Это нормализует значения высоты колонок так, что их сумма будет равна 1.

 Смысл параметра density=True^
 При density=True: Высота каждой колонки будет представлять собой плотность вероятности, то есть отношение количества значений в интервале к общему количеству значений, умноженное на ширину интервала. Это позволяет интерпретировать график как функцию плотности вероятности.
 При density=False: Высота колонок представляет собой абсолютные частоты (количество наблюдений) в каждом интервале.


```python
# Генерация случайных данных из нормального распределения
mu, sigma = 0, 2  # среднее и стандартное отклонение
data = np.random.normal(mu, sigma, 1000)  # 1000 случайных чисел

# Построение гистограммы
plt.figure(figsize=(12, 6))

# Гистограмма частот
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, density=False, alpha=0.6, color='r', label='Гистограмма')
plt.title('Гистограмма частот')
plt.xlabel('Значение')
plt.ylabel('Частота')

# Гистограмма относительных частот
plt.subplot(1, 2, 2)
plt.hist(data, bins=10, density=True, alpha=0.6, color='g', label='Гистограмма')
plt.title('Гистограмма относительных частот')
plt.xlabel('Значение')
plt.ylabel('Плотность')

plt.tight_layout()
plt.show()
```

**Пример.** Генерация нормального распределения, и построение на одном графике гистограммы относительных частот и кривой теоретического распределения с исходными параметрами

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Параметры нормального распределения
mu = 2               # математическое ожидание
sigma = 0.3          # стандартное отклонение
sample_size = 1000   # размер выборки

# Генерация выборки из нормального распределения
data = np.random.normal(mu, sigma, sample_size)

# Построение гистограммы относительных частот
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Гистограмма относительных частот')

# Построение теоретической кривой нормального распределения
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)  # диапазон x для кривой
pdf = stats.norm.pdf(x, mu, sigma)  # плотность вероятности
plt.plot(x, pdf, 'r-', label='Теоретическая кривая нормального распределения')

# Настройка графика
plt.title('Гистограмма относительных частот и теоретическая кривая')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()

# Показать график
plt.show()
```

**Примечание.** Аббревиатура **PDF** в функции stats.norm.pdf расшифровывается как Probability Density Function, что переводится как функция плотности вероятности. 

Функция плотности вероятности описывает вероятность того, что случайная величина примет определенное значение. В случае непрерывных распределений, таких как нормальное распределение, PDF определяет, насколько вероятно, что случайная величина окажется в заданном интервале значений.
Чтобы получить вероятность попадания в определенный интервал, необходимо интегрировать PDF по этому интервалу.

Таким образом, stats.norm.pdf используется для вычисления значений функции плотности вероятности нормального распределения для заданных значений.

Аббревиатура **CDF** в функции stats.norm.cdf расшифровывается как Cumulative Distribution Function, что переводится как функция распределения.

Функция распределения (CDF) описывает вероятность того, что случайная величина примет значение, меньшее заданного. То есть CDF дает вероятность того, что случайная величина будет меньше определенного значения.
Для непрерывных распределений, таких как нормальное распределение, CDF вычисляется как интеграл функции плотности вероятности (PDF) от минус бесконечности до заданного значения.

Таким образом, stats.norm.cdf используется для вычисления вероятности того, что случайная величина, следуя нормальному распределению, будет меньше заданного значения.

**Пример.** Генерация нормального распределения, и построение на одном графике кумуляты относительных частот и теоретической кривой функции распределения с исходными параметрами

```python
# Параметры нормального распределения
mu = 2               # математическое ожидание
sigma = 0.3          # стандартное отклонение
sample_size = 1000   # размер выборки

# Генерация выборки из нормального распределения
data = np.random.normal(mu, sigma, sample_size)

# Построение кумуляты относительных частот (эмпирическая функция распределения)
sorted_data = np.sort(data)
cumulative_counts = np.arange(1, sample_size + 1) / sample_size

# Построение графика
plt.step(sorted_data, cumulative_counts, label='Эмпирическая функция распределения', color='g', where='post')

# Построение теоретической кривой функции распределения
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # диапазон x для кривой
cdf = stats.norm.cdf(x, mu, sigma)  # теоретическая функция распределения 
plt.plot(x, cdf, 'r-', label='Теоретическая кривая функции распределения')

# Настройка графика
plt.title('Кумулята относительных частот и теоретическая кривая функции распределения')
plt.xlabel('Значение')
plt.ylabel('Вероятность')
plt.legend()

# Показать график
plt.grid()
plt.show()
```

**Примечание1.** Функция **np.linspace** из библиотеки NumPy используется для создания равномерно распределенной последовательности чисел. Вот основные параметры этой функции:
1. start:
   - Значение, с которого начинается последовательность. Это начальная точка интервала.

2. stop:
   - Значение, на котором заканчивается последовательность. Это конечная точка интервала. Включается в последовательность.

3. num (по умолчанию 50):
   - Количество чисел, которые нужно сгенерировать в диапазоне от start до stop. Это целое число, определяющее, сколько точек будет в результате.

4. endpoint (по умолчанию True):
   - Если True, stop включается в последовательность. Если False, stop не будет включен.

5. retstep (по умолчанию False):
   - Если True, функция возвращает кортеж, состоящий из массива и шага (разности между последовательными значениями).

6. dtype (по умолчанию None):
   - Тип данных для возвращаемого массива. Если не указан, NumPy определит тип автоматически.


**Примечание2.** Функция **plt.plot** из библиотеки Matplotlib используется для создания графиков. Вот основные параметры этой функции:

1. x:
   - Координаты по оси X. Это массив или список значений, которые будут отображены по горизонтальной оси.

4. data (по умолчанию None):
   - Опциональный аргумент, который позволяет передать данные в виде объекта, содержащего координаты. Это может быть полезно, если данные находятся в DataFrame или другом подобном формате.

5. label:
   - Метка для текущей линии, которая может использоваться в легенде графика.

6. linewidth (или lw, по умолчанию 1):
   - Толщина линии в пикселях.

7. linestyle (по умолчанию '-'):
   - Стиль линии. Можно указать различные стили, такие как '-' (сплошная линия), '--' (пунктирная линия), ':' (точечная линия) и 'None' (без линии).

8. marker:
   - Стиль маркера для отображения точек данных. Можно указать различные маркеры, такие как 'o' (круг), 's' (квадрат), '^' (треугольник) и т.д.

9. color:
   - Цвет линии или маркеров. Можно указать цвет в виде строки, например, 'red', 'blue', или использовать шестнадцатеричный код цвета.
